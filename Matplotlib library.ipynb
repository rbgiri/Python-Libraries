```python
import numpy as np
import matplotlib as mp 
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5*np.pi,10)

#defining waves:
sin_y=np.sin(x)
cos_y=np.cos(x)
#values of x and sinx:
print(x)
print([i for i in sin_y])
print([i for i in cos_y])


```

```python
x=np.linspace(0,5*np.pi,100)

#defining waves:
sin_y=np.sin(x)
cos_y=np.cos(x)

#Giving title:
plt.title("Sine Wave and Cos Wave")

#plotting the waves :
plt.plot(x,sin_y,label='Sine')
plt.plot(x,cos_y,label='Cosine')

plt.legend(loc='upper right')
plt.show()
```

```python
cars=['AUDI','BMW','FORD','SKODA','TESLA','TOYOTA']
data=[ 200  , 500 ,  240 ,  323  ,  343  ,  234]
explode=[0  ,  0  ,  0   ,0.2,0,0]

fig=plt.figure(figsize=(7,7))
plt.pie(data,labels=cars,explode=explode,shadow=True)
plt.show()
```

```python
cars=['AUDI','BMW','FORD','SKODA','TESLA','TOYOTA']
data=[ 200  , 500 ,  240 ,  323  ,  343  ,  234   ]
explode=[0  ,  0  ,   0   ,  0.1  ,   0   ,   0   ]

fig=plt.figure(figsize=(7,7))
plt.pie(data,labels=cars,explode=explode,shadow=True,autopct='%2.2f%%')
plt.show()
```

```python
fig=plt.figure(figsize=(7,7))
plt.pie(data,labels=cars,explode=explode,shadow=True,autopct='%2.2f%%',wedgeprops={'edgecolor':'black','linewidth':2,
                                                                                   'antialiased': True})
plt.show()
```

```python
x=np.linspace(0,10,50)

fig,ax=plt.subplots(2,2)

y=np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('sine wave')

y=np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title('cos wave')

y=np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title('Tan wave')

y=np.random.randn(1000)
ax[1,1].hist(y,bins=30,edgecolor='black')
ax[1,1].set_title('Random Distribution')
plt.tight_layout()
plt.show()
```

```python
plt.subplots_adjust(hspace=.4,wspace=.4)

for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(0.5,0.5,str((2,3,i)),fontsize=20,ha='center')
    plt.tight_layout()
    
```

```python
x=np.linspace(0,10,50)
y=np.sin(x)

for i in range(1,2):
    fig , (ax1,ax2,ax3)=plt.subplots(1,3,figsize=(6,6))
    
    ax1=plt.plot(x,y)
    ax2=plt.plot(x,y**2)
    ax3=plt.plot(x,y**3)
    
    ax1.set_title('Sine Wave')
    ax2.set_title('Sine Wave - Square')
    ax3.set_title('Sine Wave - Cube')
    
    plt.tight_layout()
    plt.show()
```
